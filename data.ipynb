```python
plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix, 
    annot=True,          
    fmt=".2f",           
    cmap='coolwarm',     
    vmin=-1, vmax=1, 
    square=True,      
    linewidths=.5  
)
plt.title('Heatmap Korelasi Kualitas Udara dan Cuaca', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
